```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
delivery=pd.read_csv("deliveries.csv")
match=pd.read_csv("matches.csv")
new=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
# checking

data=new[new['batsman']=='DA Warner'].groupby('season')['batsman_runs'].sum()
data
```

```python
# Q1.--function defination of the player record of per season runs--
def playerRecords(batsman):
    
    data=new[new['batsman']==batsman].groupby('season')['batsman_runs'].sum()
    plt.xlabel('season')
    plt.ylabel('runs')
    plt.title(batsman)
    plt.bar(new['season'].unique(),data)
    return plt.show()
```

```python
playerRecords('V Kohli')
```

```python
# Q2.-- Player scores % of runs with different scoring shot---
def playerScoring(batsman):
    data=new[new['batsman']==batsman].groupby('batsman_runs')['batsman_runs'].sum()
    data=data.drop(index=0) # no meaning of having runs scored in 0s. so droped it.
    plt.title(batsman)
    plt.pie(data,labels=data.index,autopct="%0.2f%%")
    return plt.show()
```

```python
playerScoring('V Kohli')
```
