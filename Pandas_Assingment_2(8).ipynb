```python
# Assingment Q1-
# 2. You have to write a function which will take 2 inputs, team 1 and team 2 and then return the win loss record between them
# By masking team-- both teams might be in column 'team1' or 'team2' -- then values count over 'winner' column gives the result.
```

```python
# Ans:-
def team1_vs_team2(team1,team2):
    m1=data['team1']==team1
    m2=data['team2']==team1
    m3=data['team1']==team2
    m4=data['team2']==team2
    return data[(m1 | m2) & (m3 | m4)]['winner'].value_counts()
team1_vs_team2('Kolkata Knight Riders','Rising Pune Supergiant')
```
